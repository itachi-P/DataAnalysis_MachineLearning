# 探索的データ分析をしていこう！

```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
files = glob.glob("train/*.csv")
```

```python
data_list = []
# 全CSVファイルを読み込み、結合
for file in files:
    data_list.append(pd.read_csv(file, index_col=0))
```

```python
df = pd.concat(data_list)
```

### data件数 0 non-null は全く意味を成さないデータなので除外対象

```python
# 「地域」などの除外対象の項目名をリストにして一括削除
nonnull_list = []
for col in df.columns:
    nonnull = df[col].count()
    if nonnull == 0:
        nonnull_list.append(col)
nonnull_list
```

```python
# 0 non-null の項目を全て除外して上書き
df = df.drop(nonnull_list, axis=1)
```

```python
df.info()
```

```python
# 「市区町村コード」と「市区町村名」は完全に一致する、分析モデルとして非常によくない（多重共線性）
df["市区町村コード"].value_counts()
```

```python
df["市区町村名"].value_counts()
```

```python
# やはり完全に一致しているので、片方を除外する
df = df.drop("市区町村名", axis=1)
```

```python
df
```

```python
# 数値型になっているデータの統計量を見てみる
df.describe()
```

```python
# 数値型でなくオブジェクト型（文字列）が入っているものの統計
df.astype("str").describe()
```

### 「最寄駅：距離（分）」や「面積（㎡）」等はobject型から数値型に変換したい

```python
# Jupyter notebookで表示されるpandasの出力結果の件数を増やしておく
pd.set_option("display.max_rows", 200)
df["最寄駅：距離（分）"].value_counts()
```

```python
# 文字の混じったデータを数値型で統一する (テストコード)
df_temp = df["最寄駅：距離（分）"].replace("30分?60分", 45)
```

```python
df_temp.value_counts()
```

```python
# 変換対象を辞書型(Python独自の配列の一種)に一旦入れて全て置き換え、更に全てfloat型に変換する
dist = {
    "30分?60分":45,
    "1H?1H30":75,
    "2H?":120,
    "1H30?2H":105
}
df["最寄駅：距離（分）"] = df["最寄駅：距離（分）"].replace(dist).astype(float)
```

```python
df["最寄駅：距離（分）"].value_counts()
```

```python
df["面積（㎡）"].value_counts()
```

```python
# 「2000㎡以上」が混じっているので数値型に置き換える
df["面積（㎡）"] = df["面積（㎡）"].replace("2000㎡以上", 2000).astype(float)
df["面積（㎡）"].value_counts()
```

### 和暦のデータを加工

```python
df["建築年"]
```

```python
df["建築年"].value_counts().keys()
```

#### まずは「戦前」以外の和暦を西暦に変換

```python
# テストコード
df["建築年"].value_counts().keys()[0].split("平成")[1].split("年")[0]
```

```python
#現在年を取得
import datetime
currentDateTime = datetime.datetime.now()
date = currentDateTime.date()
nowyear = int(date.strftime("%Y")) # 計算に使う為に数値型に型変換しておく
print(f"Current Year -> {nowyear}")
```

#### 和暦の建築年を築年数に変換する

```python
y_list = {}

for i in df["建築年"].value_counts().keys():
    if "昭和" in i:
        num = int(i.split("昭和")[1].split("年")[0])
        # year = 96 - num #この方式だとこのプログラムが元々書かれた2021年にしか正常に動作せず、毎年書き換えが必要
        year = nowyear - (1925 + num)
    if "平成" in i:
        num = int(i.split("平成")[1].split("年")[0])
        year = nowyear - (1988 + num)
    if "令和" in i:
        num = int(i.split("令和")[1].split("年")[0])
        year = nowyear - (2018 + num)
    y_list[i] = year #幅のある「戦前」には対応できないが、とりあえず並び順の問題で直前の「昭和41年」と同じ数値が入る（後で修正する）
y_list
```

```python
df["建築年"] = df["建築年"].replace(y_list)
```
