```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_list = []
# 全CSVファイルを読み込み、結合
for file in files:
    data_list.append(pd.read_csv(file, index_col=0))
```

```python
df = pd.concat(data_list)
```

### data件数 0 non-null は全く意味を成さないデータなので除外対象

```python
# 「地域」などの除外対象の項目名をリストにして一括削除
nonnull_list = []
for col in df.columns:
    nonnull = df[col].count()
    if nonnull == 0:
        nonnull_list.append(col)
nonnull_list
```

```python
df.drop(nonnull_list, axis=1)
```
